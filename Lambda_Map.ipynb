Q1 Sort all the numbers in a list in descending order using lambda function

```python
n=[23,456,57,979,4623,86]

a=sorted(n, key=lambda x:-x)
print(list(a))
```

Q2 Sort all the characters of given string using lambda function

```python
l=['Hardik','Rohit','Virat','Sky']

a=sorted(l,key=lambda x:x[0])

print(a)
```

Q3 Given list of strings, sort all the strings by last character of that string. . Use lambda function and normal function both.

```python
l=['Hardik','Rohit','Virat','Sky']

a=sorted(l,key=lambda x:x[-1])

print(a)
```

Q4 Given list of employees. This list may contain repetitions. Find all unique employee names and print them as per order of second character in that name. Use lambda function and normal function both.


```python
l=['Hardik','Rohit','Virat','Sky','Rohit','Ishan','Hardik']

a= sorted(set(l),key=lambda x:x[1])
print(a)
```

Q6 Take comma separated numbers as input from the user. Split it in list of strings. Now convert every string in this list to float using map function

```python
l=input("Enter comma seperated strings: ").split()
# l=['453','353','456','454']
# a= map(lambda x:float(x),l)

a=map(float,l)
print(list(a))
```

Q7 Given list of floating point numbers. Convert every number into string and then join all the numbers such that they are separated by pipe(|)

```python
l=[3543.0, 342.0, 342.0, 542.0, 453.0, 345.0]

l1 =list(map(str,l))
print("|".join(l1))

```

Q8 Given list of integers. Use map function to calculate square of all and store in list. Then print sum of all square values.


```python
l=[2, 13, 5, 46, 69, 3]

a=list(map(lambda x:x*x,l))
print(a)
```
