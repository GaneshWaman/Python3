Write a Python program to create a class representing a shopping cart. Include methods for adding and removing items, and calculating the total price.

```python
class shopping_cart:
  cart = []
  # def __init__(self):
  def add_item(self,add_itm):
    self.add_itm=add_itm
    shopping_cart.cart.append(self.add_itm)
    print("Adding", self.add_itm,"in cart.")
    print(shopping_cart.cart)
  def remove_item(self,remove_itm):
    try:
      self.remove_itm=remove_itm
      shopping_cart.cart.remove(self.remove_itm)
    except:
      print("No such product in cart.")
    else:
      print("Removing",self.remove_itm,"from cart.")
      print(shopping_cart.cart)
    finally:
      print("\nPlease visit again!")
```

```python
a=shopping_cart()
a.add_item('banana')
```
